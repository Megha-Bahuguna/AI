# Titanic EDA Project - Beginner

```python
# Make plots look nicer
sns.set(style="whitegrid")

# ===========================
# Step 2: Load Dataset
# ===========================
# Load Titanic dataset directly from seaborn
df = sns.load_dataset('titanic')
df.head()
```

```python
# ===========================
# Step 3: Basic Exploration
# ===========================
print("Shape of dataset:", df.shape)
print("\nColumn Info:")
print(df.info())
print("\nSummary Stats:")
print(df.describe())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ===========================
# Step 4: Univariate Analysis
# ===========================
# Target variable distribution
sns.countplot(x="survived", data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()


# Age distribution
sns.histplot(df["age"].dropna(), bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# ===========================
# Step 5: Bivariate Analysis
# ===========================
# Survival rate by Sex
sns.barplot(x="sex", y="survived", data=df)
plt.title("Survival Rate by Sex")
plt.show()

# Survival rate by Pclass
sns.barplot(x="pclass", y="survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
# ===========================
# Step 6: Handle Missing Values
# ===========================
# Fill missing Age with median
df["age"].fillna(df["age"].median(), inplace=True)

# Fill missing Embarked with mode
df["embarked"].fillna(df["embarked"].mode()[0], inplace=True)

# Add flag if Cabin info exists
df["cabin_flag"] = df["deck"].notnull().astype(int)

print("Missing values after handling:")
print(df.isnull().sum())
```

```python
# ===========================
# Step 7: Feature Engineering
# ===========================
# Family Size
df["family_size"] = df["sibsp"] + df["parch"] + 1

# Is Alone
df["is_alone"] = (df["family_size"] == 1).astype(int)

df[["family_size", "is_alone"]].head()
```

```python
# ===========================
# Step 8: Visualization Story
# ===========================
# Age vs Survival
sns.boxplot(x="survived", y="age", data=df)
plt.title("Age vs Survival")
plt.show()

# Fare vs Survival
sns.histplot(data=df, x="fare", hue="survived", bins=30, kde=True)
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
# ===========================
# Step 9: Save Clean Data
# ===========================
df.to_csv("titanic_clean.csv", index=False)
print("Clean dataset saved as titanic_clean.csv")
```
